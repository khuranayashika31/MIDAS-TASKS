### Importing required libraries

```python
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
import sys
```

```python
tokenizer=RegexpTokenizer(r'\w+')
en_stopwords=set(stopwords.words('english'))
ps=PorterStemmer()
```

### The words below are not useful in determing category. So lets discard them.

```python
notimp=['key', 'features','specifications', 'rs', 'flipkart', 'com']
```

### Function for pre-processing the description
- Converting to lower-case
- Tokenizing
- If token is found in the list above, not considering it
- Removing stop-words
- Stemming
- Removing digits

```python
def getStemmedDesc(desc):
    
    desc=desc.lower()
    for item in desc:
    # Tokenize
        tokenss=tokenizer.tokenize(desc)
        tokens=[]
        for t in tokenss:
            if t.lower() in notimp:
                pass
            else:
                tokens.append(t)
            
        new_tokens=[token for token in tokens if token not in en_stopwords]
        stemmed_tokens=[ps.stem(token) for token in new_tokens]
        cleaned_desc=' '.join(toke for toke in stemmed_tokens if not toke.isdigit())
        return cleaned_desc   
```

### Splitting training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
mod=pd.DataFrame()
```

```python
mod["desc"]=data["description"]
mod["categ"]=data["primary_category"]
```

```python
mod.head(n=10)
```

```python
x_train, x_test, y_train, y_test = train_test_split(mod["desc"], mod["categ"], test_size=0.20)
```

```python
x_train.shape
```

```python
x_train=list(x_train)
```

### Pre-processing training data

```python
goodtrain=[]
for item1 in range(len(x_train)):
    
    item=str(x_train[item1])
    a=getStemmedDesc(item)
    goodtrain.append(a)
```

```python
y_train=list(y_train)
```

### Transforming description to vector

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
x_vec=cv.fit_transform(goodtrain).toarray()
```

```python
x_vec.shape
```

### Applying  Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, GaussianNB
```

```python
mnb=MultinomialNB()
```

```python
yt=np.array(y_train)
```

```python
mnb.fit(x_vec,yt)
```

### Pre-processing testing data

```python
x_test=list(x_test)
y_test=list(y_test)
goodtest=[]
for item1 in range(len(x_test)):
    
    item=str(x_test[item1])
    b=getStemmedDesc(item)
    goodtest.append(b)
    
xt_vec=cv.transform(goodtest).toarray()
```

### Generating predictions

```python
finalAns=mnb.predict(xt_vec)
```

```python
finalAns.shape
```

```python
xt_vec.shape
```

### Measuring accuracy

```python
score=0
correct=0
wrong=0
for i in range(len(y_test)):
    if(finalAns[i]==y_test[i]):
        correct=correct+1
        score=score+1
    else:
        wrong=wrong+1
        
marks=(score/3858)*100
print(marks)
```
